```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('~/Desktop/oud_treatment_outcome/notebooks/data/merged_inputed_data.csv')

df
```

```python

# create density plot for TNT
sns.set(style="whitegrid")
ax = sns.kdeplot(df['TNT'], fill=True, color="r")
ax = sns.kdeplot(df['CNT'], fill=True, color="b")
ax.axvline(df['TNT'].mean(), color='b', linestyle='--')
plt.title('Density plot for TNT')
# legend
plt.legend(['TNT', 'CNT'])
plt.show()
```

```python

# create density plot for TNT
sns.set(style="whitegrid")
ax = sns.kdeplot(df['CNT'], fill=True, color="r")
ax.axvline(df['CNT'].mean(), color='b', linestyle='--', label='Mean')
plt.title('Density plot for CNT')
plt.legend()
plt.show()
```
