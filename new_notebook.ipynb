```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('~/Desktop/oud_treatment_outcome/notebooks/data/merged_inputed_data.csv')
```

```python
outcome = data.groupby('total_visits').agg(
                            avg_negative_tests=('TNT',np.mean),
                            total_negative_tests=('TNT',np.sum),
                            avg_consecutive_negative_tests=('CNT',np.mean),
                            responder=('responder',np.mean)
                            ).round(2).reset_index().sort_values(by='total_visits',ascending=True)

outcome
```

```python
# Plot the data
fig, ax = plt.subplots(figsize=(18, 9))
sns.set(style="whitegrid")
sns.set_context("talk")
sns.set_palette("husl")

sns.lineplot(data=outcome, x='total_visits', y='avg_negative_tests', label='Average Negative Tests', marker='o')

sns.lineplot(data=outcome, x='total_visits', y='avg_consecutive_negative_tests', label='Average Consecutive Negative Tests', marker='o')


ax.set_title('Average Negative Tests vs Average Consecutive Negative Tests')
```
