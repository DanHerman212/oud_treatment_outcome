```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
# read in 3 csv files from '../data/clean_data/' folder

# t1 == Medication Doses
t1 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/m_clean.csv')

# t2 == Self Reported Drug Use
t2 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/sru_agg.csv')

# t3 == Urine Drug Screen
t3 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/u_clean.csv')
```

```python
# merge on common columns and preserve index keys from t1 only, therefore only requiring left merge from other tables

t4 = pd.merge(t1,t2, on=['patdeid','VISIT'], how='left',).merge(t3, on=['patdeid','VISIT'], how='left')
```

```python
# check first 25 rows
display(t4.shape)

t4[:25]
```

```python
# Inspect object dtypes

t4.dtypes[t4.dtypes == 'object']
```

```python
# inspect columns

t4.columns
```

```python
# identify erroneous columns from merge
to_drop = ['Unnamed: 0_x','Unnamed: 0_y']
```

```python
# drop columns
t4.drop(to_drop, axis=1, inplace=True)
```

```python
# check shape and first 25 rows
display(t4.shape)
t4[:25]
```

We understand where the NaNs are coming from.  The sru columns only collect data on weeks 4,8,12,16,20, etc<br>

Inpute the nans to 0.0 and we are done cleaning.  Ready for feature engineering.<br>

```python
# fill the nans with 0.0
t4.fillna(0.0, inplace=True)
```

```python
# verify no more nans
t4.isnull().sum()
```

```python
# check dypes
t4.info()
```

```python
# create function to generate random sample from patdeid column
def random_patdeid():
    return np.random.choice(t4.patdeid.unique())
```

```python
# sample random patient profile to sanity check data

t4.loc[t4.patdeid == random_patdeid()][:25]
```
