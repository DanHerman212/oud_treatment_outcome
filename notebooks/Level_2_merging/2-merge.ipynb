```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
# read in 3 csv files from '../data/clean_data/' folder

# t1 == Medication Doses
t1 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/m_clean.csv')

# t2 == Self Reported Drug Use
t2 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/sru_agg.csv')

# t3 == Urine Drug Screen
t3 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/u_clean.csv')

# target == Treatment Outcome
target = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/target.csv')
```

```python
t1.shape, t2.shape, t3.shape, target.shape
```

```python
target.outcome.value_counts()
```

```python
# merge on common columns and preserve index keys from t1 only, therefore only requiring left merge from other tables

t4 = pd.merge(t1,t2, on=['patdeid'], how='left',).merge(t3, on=['patdeid'], how='left').merge(target, on=['patdeid'], how='left').reset_index(drop=True)
```

```python
t4 = t4.drop(columns=['Unnamed: 0','VISIT'])
```

```python
# inspect columns

list(t4.columns)
```

```python
# create function to generate random sample from patdeid column
def random_patdeid():
    return np.random.choice(t4.patdeid.unique())
```

```python
# sample random patient profile to check outcome

t4.loc[t4.patdeid ==random_patdeid(),['patdeid','outcome']]
```
