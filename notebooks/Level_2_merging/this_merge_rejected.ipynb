There are 3 tables we will need to merge<br>

t1 = medication<br>
t2 = self reported useM<br>
t3 = urine drug screen<br>

There is a huge problem with duplicate rows in t1 and t2 that must be addressed before merge

```python
# read medication dose data
t1 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/medication_dose.csv')
t2 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/self_reported_use.csv')
t3 = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/data/clean_data/urine_drug_screen.csv')
```

```python
# remove dupes while preserving the first row for patdeid and VISIT, to keep structure intact

t1.drop_duplicates(subset=['patdeid','VISIT'],keep='first', inplace=True)
```

```python
# new shape
t1 = t1.reset_index(drop=True)
```

```python
# accept the changes
t2.drop_duplicates(subset=['patdeid','VISIT'],keep='first', inplace=True)
```

```python
# reset index
t2.reset_index(drop=True, inplace=True)
```

```python
# merge tables to combine all data
t4 = pd.merge(t1,t2, on=['patdeid','VISIT'], how='outer').merge(t3, on=['patdeid','VISIT'], how='outer')
```
