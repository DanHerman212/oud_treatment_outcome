```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# import data
df = pd.read_csv('/Users/danherman/Desktop/oud_treatment_outcome/notebooks/Level_0_statistical_analysis/oud.csv')
```

```python
ax = df.total_visits.value_counts().sort_index(ascending=True).plot(kind='barh', figsize=(10,6), color='blueviolet', fontsize=13);
# print label on each bar with values
for p in ax.patches:
    ax.annotate(str(p.get_width()), (p.get_width() * 1.005, p.get_y() * 1.005))
```
